```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("socialmedia_networks.csv")
dataset.head()
```

# here we drop some columns for apply bionomial logistic regression -for single input (single input column)

```python
dataset.drop( columns=["User_ID","Time_Spent_Minutes","Posts_Per_Week","Likes_Per_Post","Platform"],inplace=True)
```

```python
dataset.head()
```

lets  cheking this data is applicable for logistic regression or not -with making graph

```python
sns.scatterplot(x="Age", y="Clicked_Ad",data=dataset)
plt.show()
```

```python
#yes this data is applicable for logistic regression
```

lets divide the data into dependend and independent data , and split futher in train ans testing

```python
x=dataset[["Age"]]
y=dataset["Clicked_Ad"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100
```

# now lets predict that age person has purchase the click_ad or not

```python
lr.predict([[25]]) #put the age number in sq bracket
```

# lets see idea of how prediction line made

```python
sns.scatterplot(x="Age", y="Clicked_Ad",data=dataset)
sns.lineplot(x="Age",y=lr.predict(x),data=dataset,color="red")
plt.show()
```
