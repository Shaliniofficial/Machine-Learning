```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Read Dataset</h2>

```python
dataset=pd.read_csv("marketing_roi_dataset.csv")
dataset.head()
```

<h2>Data Preprocessing</h2>

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
dataset["Gender"]=le.fit_transform(dataset["Gender"])
dataset["Location "]=le.fit_transform(dataset["Location"])
dataset["Income_Level"]=le.fit_transform(dataset["Income_Level"])
dataset["Channel"]=le.fit_transform(dataset["Channel"])
dataset["Campaign_Name"]=le.fit_transform(dataset["Campaign_Name"])
```

<h2>Visualization</h2>

```python
<h3>roi distribution
```

```python
plt.figure(figsize=(10,6))
plt.hist(dataset["ROI (%)"],bins=30, edgecolor="black")
plt.xlabel("ROI (%)")
plt.ylabel("Frequency")
plt.show()
```
