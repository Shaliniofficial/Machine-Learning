```python
import pandas as pd
```

```python
df= pd.DataFrame({"size":["s","m","l","xl","xxl","s","xxxl","m","s","l","xl"]})
df.head()
```

```python
Ord_data=[["s","m","l","xl","xxl","xxxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
Oe=OrdinalEncoder(categories=Ord_data)
Oe.fit(df[["size"]])
```

```python
Oe.transform(df[["size"]])
```

```python
df["size_en"]=Oe.transform(df[["size"]])
df
```

```python
#encoding throiugh map function
```

```python
Ord_data1={"s":0,"m":1,"l":2,"xl":3,"xxl":6,"xxxl":5}
```

```python
df["size"].map(Ord_data1)
```

```python
df["size_en_map"]=df["size"].map(Ord_data1)
```
