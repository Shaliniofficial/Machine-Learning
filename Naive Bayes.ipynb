```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("CGPA_Placement_Data.csv")
```

```python
df.head()
```

<h3> lets check the data is normal distribution or not</h3>

```python
sns.kdeplot(data=df["CGPA"])
```

```python
sns.kdeplot(data=df["Score"])
```

<h3>here the both curve is normal distribution</h3>

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(4,3))

sns.scatterplot(x="CGPA",y="Score",data=df,hue="Placed")
plt.show()
```

```python
x=df[["CGPA","Score"]]
```

```python
y=df["Placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

<h4>gaussian NB</h4>

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)*100,gnb.score(x_train,y_train)*100
```

```python
y_predict=gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) * 100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

<h4>multinomial NB</h4>

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)*100,mnb.score(x_train,y_train)*100
```

```python
y_predict=mnb.predict(x_test)
accuracy_score(y_test, y_predict) * 100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

<h4>BernoulliNB</h4>

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)*100,bnb.score(x_train,y_train)*100
```

```python
y_predict=bnb.predict(x_test)
accuracy_score(y_test, y_predict) * 100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```

<h4> from all of these we get that here we use gauussain NB model because there curvature is perfect</h4>
