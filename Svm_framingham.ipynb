```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
dataset=pd.read_csv("framingham.csv")
dataset.head()
```

```python
dataset.info()
```

<h2>data preprocessing</h2>

```python
dataset.isnull().sum()
```

```python
dataset.fillna(dataset.mean(),inplace=True)
```

<h2>feature selection</h2>

```python
x=dataset.drop(columns=["TenYearCHD"])
```

```python
type(x)
```

```python
y=dataset["TenYearCHD"]
```

<h2>data validation</h2>

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
scaler=StandardScaler() #1 standard deviation,#2  then scaler matrix banenge ,#3. dot product karenge
```

<h2>scale the training and testing feature</h2>

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
#kernal=rbf,sgmoid,poly           #sigmoid used in classification liner data
svm_model=SVC(kernel="poly",class_weight="balanced")# balenced weight = apne ap value ko balence rakhega
svm_model.fit(x_train_scaled,y_train)
```

<h2> model_fitting</h2>

```python
y_predict=svm_model.predict(x_test_scaled)
```

```python
conf_matrix=confusion_matrix(y_test,y_predict)
print("\nconfusion matrix-\n",conf_matrix)
class_report=classification_report(y_test,y_predict)
print(" \nclassification_report-\n",class_report)
```
