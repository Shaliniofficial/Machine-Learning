```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\shalini maurya\Downloads\loan_data_set.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["ApplicantIncome"].fillna(dataset["ApplicantIncome"].mean())
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["ApplicantIncome"])
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["ApplicantIncome"]])
```

```python
ss.transform(dataset[["ApplicantIncome"]])
```

```python
dataset["new_ApplicantIncome_ss"]=pd.DataFrame(ss.transform(dataset[["ApplicantIncome"]]),columns=["x"])
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("before")
sns.distplot(dataset["ApplicantIncome"])

plt.subplot(1,2,2)
plt.title("after")
sns.distplot(dataset["new_ApplicantIncome_ss"])
            
```
